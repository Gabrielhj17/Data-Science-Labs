### 2e) What is the _interquartile-range_ of the number of births per day for each day of the week?
*You might need to google this concept: good practice at real life...!*. This [video](https://www.youtube.com/watch?v=qNKOi08NxHs) is helpful.

```python
# Convert 'date' column to datetime format
df2['date'] = pd.to_datetime(df2['date'])

# Extract the day of the week and add it as a new column
df2['day_of_week'] = df2['date'].dt.day_name()  # This gives the full name of the day

# Group by the day of the week and calculate the interquartile range (IQR)
iqr_births_per_day = df2.groupby('day_of_week')['births'].agg(lambda x: x.quantile(0.75) - x.quantile(0.25))

# Sort the index to maintain the order of days
iqr_births_per_day = iqr_births_per_day.reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
)

print(iqr_births_per_day)
```

### 2f) Find the _outliers_ for the number of births per day for each day of the week. Use the interquartile range method shown [here](https://online.stat.psu.edu/stat200/lesson/3/3.2), where
1. We denote the interquartile range $IQR$, and the quartile boundaries as $Q_1, Q_2$  and $Q_3$.
2. We measure outliers as anything below $Q_1 - 1.5IQR$, or above $Q_3 + 1.5IQR$

```python
# Convert 'date' column to datetime format
df2['date'] = pd.to_datetime(df2['date'])

# Extract the day of the week and add it as a new column
df2['day_of_week'] = df2['date'].dt.day_name()  # This gives the full name of the day

# Function to find outliers for each day of the week
def find_outliers(group):
    Q1 = group.quantile(0.25)
    Q3 = group.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return group[(group < lower_bound) | (group > upper_bound)]

# Group by the day of the week and find outliers
outliers_per_day = df2.groupby('day_of_week')['births'].apply(find_outliers)

# Drop any NaN values to clean up the output
outliers_per_day = outliers_per_day.dropna()

print(outliers_per_day)
```

### 2i) Find by what day of the year _50%_ of births have occured for each year.  

```python
# Convert 'date' column to datetime format
df2['date'] = pd.to_datetime(df2['date'])

# Extract year and day of the year
df2['year'] = df2['date'].dt.year
df2['day_of_year'] = df2['date'].dt.dayofyear

# Group by year and day_of_year, and calculate the sum of births
daily_births = df2.groupby(['year', 'day_of_year'])['births'].sum().reset_index()

# Calculate cumulative sum of births for each year
daily_births['cumulative_births'] = daily_births.groupby('year')['births'].cumsum()

# Calculate total births for each year
total_births_per_year = daily_births.groupby('year')['births'].sum().reset_index()
total_births_per_year.rename(columns={'births': 'total_births'}, inplace=True)

# Merge total births back to the daily_births DataFrame
daily_births = daily_births.merge(total_births_per_year, on='year')

# Calculate the threshold for 50% of total births
daily_births['threshold'] = daily_births['total_births'] * 0.5

# Find the day of the year when cumulative births exceed the threshold
day_50_percent = daily_births[daily_births['cumulative_births'] >= daily_births['threshold']]

# Get the first occurrence of the day when 50% of births have occurred
day_50_percent = day_50_percent.groupby('year').first().reset_index()

# Convert day_of_year to actual date
day_50_percent['date_at_50%'] = pd.to_datetime(day_50_percent['day_of_year'], format='%j',
                                                 errors='coerce').dt.strftime('%Y-%m-%d')

# Rename the columns for clarity
day_50_percent.rename(columns={
    'day_of_year': 'day_of_year_50%',
    'cumulative_births': 'cumulative_births_at_50%'
}, inplace=True)

print(day_50_percent[['year', 'day_of_year_50%', 'date_at_50%', 'cumulative_births_at_50%']])
```

### 2j) In 2000, _between_ what days of the year do 50% of births occur?

```python
# Convert 'date' column to datetime format
df2['date'] = pd.to_datetime(df2['date'])

# Extract year and day of the year
df2['year'] = df2['date'].dt.year
df2['day_of_year'] = df2['date'].dt.dayofyear

# Filter data for the year 2000
births_2000 = df2[df2['year'] == 2000]

# Group by day_of_year and calculate the sum of births
daily_births_2000 = births_2000.groupby('day_of_year')['births'].sum().reset_index()

# Calculate total births for the year 2000
total_births_2000 = daily_births_2000['births'].sum()

# Calculate cumulative sum of births
daily_births_2000['cumulative_births'] = daily_births_2000['births'].cumsum()

# Calculate the threshold for 50% of total births
threshold = total_births_2000 * 0.5

# Find the range of days of the year where cumulative births exceed the threshold
days_above_threshold = daily_births_2000[daily_births_2000['cumulative_births'] >= threshold]

# Get the first and last days when cumulative births exceed the threshold
start_day = days_above_threshold['day_of_year'].iloc[0]
end_day = days_above_threshold['day_of_year'].iloc[-1]

# Convert day_of_year back to dates
start_date = pd.to_datetime(f'2000-01-01') + pd.to_timedelta(start_day - 1, unit='D')
end_date = pd.to_datetime(f'2000-01-01') + pd.to_timedelta(end_day - 1, unit='D')

print(f"In 2000, 50% of births occurred between {start_date.strftime('%Y-%m-%d')} and {end_date.strftime('%Y-%m-%d')} (days {start_day} to {end_day}).")
```

### 2k) What about all the other years?

```python
# Convert 'date' column to datetime format
df2['date'] = pd.to_datetime(df2['date'])

# Extract year and day of the year
df2['year'] = df2['date'].dt.year
df2['day_of_year'] = df2['date'].dt.dayofyear

# Group the data by year and day_of_year, and calculate the sum of births
daily_births = df2.groupby(['year', 'day_of_year'])['births'].sum().reset_index()

# Initialize an empty list to store results
results = []

# Loop through each year in the dataset
for year in daily_births['year'].unique():
    # Filter data for the specific year
    births_year = daily_births[daily_births['year'] == year].copy()  # Make a copy to avoid SettingWithCopyWarning

    # Calculate total births for that year
    total_births = births_year['births'].sum()

    # Calculate cumulative sum of births
    births_year.loc[:, 'cumulative_births'] = births_year['births'].cumsum()  # Use .loc to avoid warning

    # Calculate the threshold for 50% of total births
    threshold = total_births * 0.5

    # Find the days of the year where cumulative births exceed the threshold
    days_above_threshold = births_year[births_year['cumulative_births'] >= threshold]

    # Get the first and last days when cumulative births exceed the threshold
    if not days_above_threshold.empty:
        start_day = days_above_threshold['day_of_year'].iloc[0]
        end_day = days_above_threshold['day_of_year'].iloc[-1]

        # Convert day_of_year back to dates
        start_date = pd.to_datetime(f'{year}-01-01') + pd.to_timedelta(start_day - 1, unit='D')
        end_date = pd.to_datetime(f'{year}-01-01') + pd.to_timedelta(end_day - 1, unit='D')

        # Append results for the current year
        results.append({
            'year': year,
            'start_date': start_date.strftime('%Y-%m-%d'),
            'end_date': end_date.strftime('%Y-%m-%d'),
            'days_range': f'Days {start_day} to {end_day}'
        })

# Create a DataFrame from the results
results_df = pd.DataFrame(results)

# Print the results
print(results_df)
```
